1. fitting it to the periodic function

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import optimize

max_data = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]) # Assigning above data to max_data
min_data = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]) # Assigning above data to min_data
months = np.arange(12)                                                # months is an array which contains 12 values
plt.plot(months, max_data, 'go') # Plot months on X-axis and max_data on Y-axis with red circles
plt.plot(months, min_data, 'bo') # Plot months on X-axis and min_data on Y-axis with blue circles
plt.xlabel('Months')
plt.ylabel('Temperature')
```

```python
def temperature(times, avg, height, offset):
    return (avg + height * np.cos((times + offset) * 1.8 * np.pi / times.max()))

result_max, cov_max = optimize.curve_fit(temperature, months,max_data, [40, 20, 0]) 
result_min, cov_min = optimize.curve_fit(temperature, months,min_data, [-40, 20, 0])
```

Plotting the fit

```python
days = np.linspace(0, 12, num=365)
plt.plot(months, max_data, 'go')
plt.plot(days, temperature(days, *result_max), 'g-')
plt.plot(months, min_data, 'bo')
plt.plot(days, temperature(days, *result_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
```

```python
import pandas as pd
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

Create a pie chart presenting the male/female proportion

```python
titanic.sex.value_counts().plot(kind='pie',autopct='%1.2f%%')
```

Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
titanic.sex.dropna(inplace=True)
mapping = {'male' : 'blue', 'female' : 'red'}
plt.scatter(titanic['age'], titanic['fare'], alpha=0.4, c=titanic['sex'].map(mapping))
```
